```python
##Link to dataset on website
##https://data.wprdc.org/dataset/city-of-pittsburgh-courts/resource/a5b71bfa-840c-4c86-8f43-07a9ae854227

##Imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

courtsData = pd.read_csv("courts_and_rinks.csv")
##This is what the data looks like
courtsData.head(10)
```

```python
#Groups each court or rink by which neighborhood it is found in. Disregards the type and just adds them all up
courtsGroup=courtsData.groupby("neighborhood").count()
#Drops all 'unnecessary' columns. Setting the data to a new variable
courtsSlim=courtsGroup.drop(columns=['park','name','id','location','surface_material','grandstand','inactive','council_district','ward','tract','public_works_division','pli_division','police_zone', 'fire_zone', 'latitude','longitude'])
courtsSlim.head(50)
```

```python
#Remove all neighborhoods with 2 or less in the quantity section
courtsFinal=courtsSlim.query('type>1')
courtsFinal.head(50)
```

```python
##courtsFinal.plot.bar()
```

```python
##Seperate the neighborhoods by number of courts. Points are given for the number of courts or rinks in the neighborhood
##5 points are given to any neighborhood with more than 20 courts/rinks
courts5 = courtsFinal[courtsFinal['type']>19]
fullcourts5 = courts5.assign(score=5)
print(fullcourts5)
```

```python
##4 points are given to any neighborhood with 10 or more and less than 20 courts/rinks
courts4 = courtsFinal[(courtsFinal['type']>9) & (courtsFinal['type']<20)]
fullcourts4 = courts4.assign(score=4)
print(fullcourts4)
```

```python
##3 points are given to any neighborhood with 7 or more and less than 10 courts/rinks
courts3 = courtsFinal[(courtsFinal['type']>7) & (courtsFinal['type']<10)]
fullcourts3 = courts3.assign(score=3)
print(fullcourts3)
```

```python
##2 points are given to any neighborhood with 5 or more and less than 7 courts/rinks
courts2 = courtsFinal[(courtsFinal['type']>4) & (courtsFinal['type']<8)]
fullcourts2 = courts2.assign(score=2)
print(fullcourts2)
```

```python
##1 point is given to any neighborhood with 3 or more and less than 5 courts/rinks
courts1 = courtsFinal[(courtsFinal['type']>1) & (courtsFinal['type']<5)]
fullcourts1 = courts1.assign(score=1)
print(fullcourts1)
```
