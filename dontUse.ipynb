**Obesity Rates (Sean)**

```python
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt

obesity_data = pd.read_csv("ObesityDataSet.csv")
obesity_data.head(10)
```

First, I imported the dataset regarding obesity in Pittsburgh. The essence of my metric is that the neighborhood with the least amount of obesity would be the best place to live. No, this is not a fatphobic concept, I am simply choosing which neighborhood appears to be the healthiest by their obesity rates.

```python
obesity_data_condensed = obesity_data.drop(labels=['2000 Tract', '2010 Tract', 'MUNICIPALITY'], axis = 1)
obesity_data_condensed.head(10)
```

I then trimmed the data by removing uneccesary columns. The columns removed would not have had any effect on the health level, they were just identifiers for the neighborhood in which the data was collected.

```python
obesity_values = obesity_data_condensed.sort_values('2006-2010 estimate of obesity', ascending = True)
obesity_values.head(10)
```

```python
query_mask = obesity_values['2006-2010 estimate of obesity'] <= 0.165
condensed_values = obesity_values[query_mask]
condensed_values.head(10)
```

The rates of obesity were then put in order. The table was put in ascending order meaning the value at the top was the neighborhood with the lowest rate of obesity.

```python
condensed_values.plot(x='City Neighborhood', y='2006-2010 estimate of obesity', kind='bar', figsize=(15,5))
```

This graph shows the neighborhoods with the lowest estimate of obesity

**Number of Basketball Courts (Philippe)**

```python
##Link to dataset on website
##https://data.wprdc.org/dataset/city-of-pittsburgh-courts/resource/a5b71bfa-840c-4c86-8f43-07a9ae854227

##Imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

courtsData = pd.read_csv("courts_and_rinks.csv")
##This is what the data looks like
courtsData.head(10)
```

```python
#Groups each court or rink by which neighborhood it is found in. Disregards the type and just adds them all up
courtsGroup=courtsData.groupby("neighborhood").count()
#Drops all 'unnecessary' columns. Setting the data to a new variable.
courtsFinal=courtsGroup.drop(columns=['park','name','id','location','surface_material','grandstand','inactive','council_district','ward','tract','public_works_division','pli_division','police_zone', 'fire_zone', 'latitude','longitude'])
##Here is a visualization for what the data set looks like right now
courtsFinal.head(10)
```

```python
##Seperate the neighborhoods by number of courts. Points are given for the number of courts or rinks in the neighborhood
##5 points are given to any neighborhood with more than 20 courts/rinks
courts5 = courtsFinal[courtsFinal['type']>19]
fullcourts5 = courts5.assign(score=5)
##These are the neighborhoods that receive 5 pts
print(fullcourts5)
```

```python
##4 points are given to any neighborhood with 10 or more and less than 20 courts/rinks
courts4 = courtsFinal[(courtsFinal['type']>9) & (courtsFinal['type']<20)]
fullcourts4 = courts4.assign(score=4)
##These are the neighborhoods that receive 4 pts
print(fullcourts4)
```

```python
##3 points are given to any neighborhood with 7 or more and less than 10 courts/rinks
courts3 = courtsFinal[(courtsFinal['type']>7) & (courtsFinal['type']<10)]
fullcourts3 = courts3.assign(score=3)
##These are the neighborhoods that receive 3 pts
print(fullcourts3)
```

```python
##2 points are given to any neighborhood with 5 or more and less than 7 courts/rinks
courts2 = courtsFinal[(courtsFinal['type']>4) & (courtsFinal['type']<8)]
fullcourts2 = courts2.assign(score=2)
##These are the neighborhoods that receive 2 pts
print(fullcourts2)
```

```python
##1 point is given to any neighborhood with less than 5 courts/rinks
courts1 = courtsFinal[(courtsFinal['type']<5)]
fullcourts1 = courts1.assign(score=1)
##These are the neighborhoods that receive 1 point
print(fullcourts1)
```

**Median Death Age (Charles)**

In order to determine which neighborhood in Pittsburgh is the healthiest, it is important for us to look at the median age of death. This tells us which place is the best for living a full, healthy life. A higher median age of death means that the neighborhood is more likely to have better healthcare and lifestyles.

This metric will involve awarding points on a scale of 1 to 5. The worst neighborhood gets 1 point and the best gets 5 points. The median neighborhood will get 3 points. All other neighborhoods will either get 2 or 4 points depending on where they fall.

```python
##Link to dataset on website
##https://data.wprdc.org/dataset/median-age-death

##Imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

pd.read_csv("DeathAge.csv")
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

df=pd.read_csv("DeathAge.csv")
maximum=df['TOTAL MD AGE AT DEATH'].max()
print("The maximum median age is: " + str(maximum))
minimum=df['TOTAL MD AGE AT DEATH'].min()
print("The minimum median age at death is: " + str(minimum))
med= df['TOTAL MD AGE AT DEATH'].median()
print("The median median age at death is: " + str(med))
```

```python
import pandas as pd
best_place = pd.read_csv("DeathAge.csv")
query_mask = best_place['TOTAL MD AGE AT DEATH'] == 85.9
winner=best_place[query_mask]
print("The best neighborhood (that will receive 5 points) is: ")
winner.head(10)
```

```python
import pandas as pd
worst_place = pd.read_csv("DeathAge.csv")
query_mask = worst_place['TOTAL MD AGE AT DEATH'] == 53.3
loser=best_place[query_mask]
print("The worst neighborhood (that will receive 1 point) is: ")
loser.head(10)
```

```python
import pandas as pd
medium_place = pd.read_csv("DeathAge.csv")
query_mask = medium_place['TOTAL MD AGE AT DEATH'] == 76.6
medium=medium_place[query_mask]
print("The most medium neighborhood (that will receive 3 points) is: ")
medium.head(10)
```

```python
import pandas as pd
mediocre_places = pd.read_csv("DeathAge.csv")
query_mask = (mediocre_places['TOTAL MD AGE AT DEATH']>53.3) & (mediocre_places['TOTAL MD AGE AT DEATH']<76.6)
mediocre=mediocre_places[query_mask]
print("The most mediocre neighborhoods (that will receive 2 points) are: ")
mediocre.head(90)
```

```python
import pandas as pd
great_places = pd.read_csv("DeathAge.csv")
query_mask = (great_places['TOTAL MD AGE AT DEATH']>76.6) & (great_places['TOTAL MD AGE AT DEATH']<85.9)
great=great_places[query_mask]
print("The great neighborhoods (that will receive 4 points) are: ")
great.head(100)
```
