**Median Age of Death in Pittsburgh Introduction**

 In order to determine which neighborhood in Pittsburgh is the healthiest, it is important for us to look at the median age of death. This tells us which place is the best for living a full, healthy life. A higher median age of death means that the neighborhood is more likely to have better healthcare and lifestyles.

**Metric**

Our metric will involve awarding points on a scale of 1 to 5. The worst neighborhood gets 1 points and the best gets 5 points. The median neighborhood will get 3 points. All other neighborhoods will either get 2 or 3 points depending on where they fall.

```python
##Link to dataset on website
##https://data.wprdc.org/dataset/median-age-death

##Imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

pd.read_csv("DeathAge.csv")
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

df=pd.read_csv("DeathAge.csv")
maximum=df['TOTAL MD AGE AT DEATH'].max()
print("The maximum median age is: " + str(maximum))
minimum=df['TOTAL MD AGE AT DEATH'].min()
print("The minimum median age at death is: " + str(minimum))
med= df['TOTAL MD AGE AT DEATH'].median()
print("The median median age at death is: " + str(med))
```

```python
import pandas as pd
best_place = pd.read_csv("DeathAge.csv")
query_mask = best_place['TOTAL MD AGE AT DEATH'] == 85.9
winner=best_place[query_mask]
print("The best neighborhood (that will receive 5 points) is: ")
winner.head(10)
```

```python
import pandas as pd
worst_place = pd.read_csv("DeathAge.csv")
query_mask = worst_place['TOTAL MD AGE AT DEATH'] == 53.3
loser=best_place[query_mask]
print("The worst neighborhood (that will receive 1 point) is: ")
loser.head(10)
```

```python
import pandas as pd
medium_place = pd.read_csv("DeathAge.csv")
query_mask = medium_place['TOTAL MD AGE AT DEATH'] == 76.6
medium=medium_place[query_mask]
print("The most medium neighborhood (that will receive 3 points) is: ")
medium.head(10)
```

```python
import pandas as pd
mediocre_places = pd.read_csv("DeathAge.csv")
query_mask = (mediocre_places['TOTAL MD AGE AT DEATH']>53.3) & (mediocre_places['TOTAL MD AGE AT DEATH']<76.6)
mediocre=mediocre_places[query_mask]
print("The most mediocre neighborhoods (that will receive 2 points) are: ")
mediocre.head(90)
```

```python
import pandas as pd
great_places = pd.read_csv("DeathAge.csv")
query_mask = (great_places['TOTAL MD AGE AT DEATH']>76.6) & (great_places['TOTAL MD AGE AT DEATH']<85.9)
great=great_places[query_mask]
print("The great neighborhoods (that will receive 4 points) are: ")
great.head(100)
```
